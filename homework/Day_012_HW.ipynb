```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
boston_choice = pd.read_csv('boston.csv', usecols=['NOX','DIS'])
boston_choice
```

```python
#散佈圖
df = pd.DataFrame(boston_choice, columns=["NOX","DIS"])
print(df)
df.plot.scatter(x='NOX', y='DIS',figsize=(10,8))
```

```python
print('以散布圖推斷"NOX"和"DIS"兩者大致上呈現反比關係')
```
